```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data.head()
```

```python
# calculation
total_players = purchase_data["SN"].nunique()
#data Frame 
data = {"Total Players":[total_players]}
total_players_df = pd.DataFrame.from_dict(data)
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculations 

unique_items = purchase_data["Item ID"].nunique()
#unique_item=len(purchase_data["Item ID"].unique())

avg_price = purchase_data["Price"].mean() #.round(2)

nbr_purchases = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

#dataframe 

data_analysis = {"Number of Unique Items":[unique_items],"Average Price":avg_price,\
        "Number of purchases":nbr_purchases, "Total Revenue":total_revenue} 
DataAnalysis_df = pd.DataFrame.from_dict(data_analysis)
DataAnalysis_df['Average Price'] = DataAnalysis_df['Average Price'].map("${:,.2f}".format)
DataAnalysis_df['Total Revenue'] = DataAnalysis_df['Total Revenue'].map("${:,.2f}".format)
DataAnalysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# total count by gender 
total_count = purchase_data.groupby("Gender").SN.nunique()

#percentage of total count by gender 
percentage_count = ((total_count/total_players)*100).round(2)

#DataFrame 
dic = {"Total Count": total_count, "Percentage of Players":percentage_count}
total_count_df = pd.DataFrame(dic)
total_count_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchase count by gender 
purchase_count = purchase_data.groupby("Gender")["Purchase ID"].count()

# Average purchase Price by gender 
avg_purchase_price = purchase_data.groupby("Gender")["Price"].mean()

#total purchase values by gender 
total_purchase_value = purchase_data.groupby("Gender")["Price"].sum()

# avg Total Purchase per Person
avg_total_price_per_person =total_purchase_value/total_count_df["Total Count"]

#DataFrame 
dic = {"Purchase Count":purchase_count, "Average Purchase Price":avg_purchase_price, "Total Purchase Value":total_purchase_value,\
   "Avg Total Purchase per Person":avg_total_price_per_person}
purchasing_analysis_df = pd.DataFrame(dic)

#Data Formating 
purchasing_analysis_df["Total Purchase Value"] =purchasing_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_df["Avg Total Purchase per Person"] = purchasing_analysis_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchasing_analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 



```

```python
#calculations of Total count & Percentage of Players & round the percentage 
total_by_age = purchase_data.groupby('Age')['SN'].nunique()
percent_of_players_by_age = ((total_by_age/total_players)*100).round(2)

#dataFrame 
dic = {"Total Count":total_by_age, "Percentage of Players":percent_of_players_by_age}
age_demographics_df = pd.DataFrame(dic)


#set age column as datafram 
age_demographics_df['Age'] = age_demographics_df.index



 # Place the Age groups into a new column inside of age_demographics_df 
age_demographics_df["Age Groups"] =  pd.cut(age_demographics_df["Age"], bins, labels=group_labels)
age_demographics_df = age_demographics_df.groupby("Age Groups").sum()

#del age column 
del age_demographics_df['Age']

#final dataframe 

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:,.2f} %".format)

age_demographics_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
 # Place the Age groups into a new column inside of age_demographics_df 
purchase_data["Age Groups"] =  pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()
```

```python
#calculations 
group_age = purchase_data.groupby("Age Groups")

purchase_count = group_age["Purchase ID"].count()
avg_purchase_price = group_age["Price"].mean()
total_purchase_value = group_age["Price"].sum()


#dataFrame 
dic = {"Purchase Count":purchase_count,
       "Average Purchase Price":avg_purchase_price,
       "Total Purchase Value":total_purchase_value}
#        "Avg Total Purchase per Person":avg_total_purchase_per_person
purchasing_analysis_by_age_df = pd.DataFrame(dic)

#give the displayed data cleaner formatting
purchasing_analysis_by_age_df["Average Purchase Price"] = purchasing_analysis_by_age_df["Average Purchase Price"].\
                                                        map("${:,.2f}".format)
purchasing_analysis_by_age_df["Total Purchase Value"] = purchasing_analysis_by_age_df["Total Purchase Value"].\
                                                        map("${:,.2f}".format)

purchasing_analysis_by_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#calculations 
group_sn = purchase_data.groupby("SN")

purchase_count = group_sn["Purchase ID"].count()
avg_purchase_price = group_sn["Price"].mean()
total_purchase_value = group_sn["Price"].sum()


#dataFrame 
dic = {"Purchase Count":purchase_count,
       "Average Purchase Price":avg_purchase_price,
       "Total Purchase Value":total_purchase_value}


Top_spenders_df= pd.DataFrame(dic)

#Sort the total purchase value column in descending order
Top_spenders_df = Top_spenders_df.sort_values('Total Purchase Value', ascending=False)
Top_spenders_df["Total Purchase Value"]= Top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)
Top_spenders_df["Average Purchase Price"]= Top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)

Top_spenders_df.head()
```
